```python
import numpy as np
import pandas as pd
import statistics as stats

from scipy.spatial.distance import pdist, squareform
```

Exercice 1_1_1

```python
means = np.array([15, 500, 5, 60, 1013, 4])
deviation = np.array([7.5, 200, 2.5, 15.0, 10.0, 2.0])

matrix = np.eye(6)

matrix[0, 1] = matrix[1, 0] = 0.6
matrix[0, 2] = matrix[2, 0] = -0.5
matrix[3, 2] = matrix[2, 3] = 0.6
matrix[5, 1] = matrix[1, 5] = 0.7
matrix[5, 2] = matrix[2, 5] = -0.4

diag = np.diag(deviation)
sigma = diag @ matrix @ diag

data = np.random.multivariate_normal(means, sigma, size=300)

dataFrame = pd.DataFrame(data, columns=['Temp', 'Sun', 'Rain', 'Hum', 'Press', 'UV'])

dataFrame['Rain'] = dataFrame['Rain'].clip(lower=0)
dataFrame['Sun'] = dataFrame['Sun'].clip(lower=0)
dataFrame['UV'] = dataFrame['UV'].clip(lower=0)

dataFrame['Hum'] = dataFrame['Hum'].round().astype(int)
dataFrame['UV'] = dataFrame['UV'].round().astype(int)

dataFrame.to_csv('artificial_dataset.csv', index=False)
```

Exercice 1_1_2

```python
dataset = pd.read_csv('artificial_dataset.csv')

not_normalized_data = dataset.to_numpy()
# z-score (Standard Scaler)
normalized_df = (dataset - dataset.mean()) / dataset.std()
normalized_data = normalized_df.to_numpy()

not_normalized_data_dst = pdist(not_normalized_data, "euclidean")
not_normalized_data_sf = squareform(not_normalized_data_dst)
np.fill_diagonal(not_normalized_data_sf, np.infty)

normalized_data_min = np.unravel_index(np.argmin(not_normalized_data_sf), not_normalized_data_sf)
normalized_data_max = np.unravel_index(np.argmax(not_normalized_data_sf), not_normalized_data_sf)

normalized_data_dst = pdist(normalized_data, "euclidean")
normalized_data_sf = squareform(normalized_data_dst)
np.fill_diagonal(normalized_data_sf, np.infty)

normalized_data_min = np.unravel_index(np.argmin(normalized_data_sf), normalized_data_sf)
normalized_data_max = np.unravel_index(np.argmax(normalized_data_sf), normalized_data_sf)
```

```python
means = np.array([15, 500, 5, 60, 1013, 4])
deviation = np.array([7.5, 200, 2.5, 15.0, 10.0, 2.0])

matrix = np.eye(6)

matrix[0, 1] = matrix[1, 0] = 0.6
matrix[0, 2] = matrix[2, 0] = -0.5
matrix[3, 2] = matrix[2, 3] = 0.6
matrix[5, 1] = matrix[1, 5] = 0.7
matrix[5, 2] = matrix[2, 5] = -0.4

diag = np.diag(deviation)
sigma = diag @ matrix @ diag

data = np.random.multivariate_normal(means, sigma, size=300)

dataFrame = pd.DataFrame(data, columns=['Temp', 'Sun', 'Rain', 'Hum', 'Press', 'UV'])

dataFrame['Rain'] = dataFrame['Rain'].clip(lower=0)
dataFrame['Sun'] = dataFrame['Sun'].clip(lower=0)
dataFrame['UV'] = dataFrame['UV'].clip(lower=0)

dataFrame['Hum'] = dataFrame['Hum'].round().astype(int)
dataFrame['UV'] = dataFrame['UV'].round().astype(int)

dataFrame.to_csv('artificial_dataset.csv', index=False)
```
